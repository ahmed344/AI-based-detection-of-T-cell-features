```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from PIL import Image
from skimage.external.tifffile import imsave
import os
import fnmatch

# Functions
def radial_profile(data):
    
    y, x = np.indices((data.shape))
    
    center = np.array([(x.max()-x.min())/2.0, (y.max()-y.min())/2.0])
    
    r = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    r = r.astype(np.int)

    tbin = np.bincount(r.ravel(), data.ravel())
    nr = np.bincount(r.ravel())
    radialprofile = tbin / nr
    return radialprofile 
```

```python
# Read the stack
Stack = io.imread('data/actin_comparison/data_20190208_cell_1_1_pattern_dots.tif')

for img,i in zip(Stack, range(Stack.shape[0])):
    # Take the radial profile
    radialprofile = radial_profile(img)
    
    plt.figure(figsize=(15,10))
    plt.plot(radialprofile, label= '{}'.format(i))
    plt.xlabel('Radial [0.1 micron]', fontsize = 'x-large')
    plt.ylabel('Intensity', fontsize = 'x-large')
    plt.xticks(fontsize = 'x-large')
    plt.yticks(fontsize = 'x-large')
    plt.grid()
    plt.legend(fontsize = 'x-large')
    plt.title('Radial profile of the pattern', fontsize = 'xx-large')
    plt.savefig('data/actin_comparison/data_20190208_cell_1_1/{}_pattern_profile'.format(i))
    plt.show()
```
