## Quantification of the size(FWHM) and contrast of the antibody nano-dots from fluorescence images of the underlying Neutravidin pattern.
 

```python
import matplotlib.pyplot as plt
```

### Quantification of the size

```python
plt.figure(figsize=(10,10))

bp = plt.boxplot([[896, 642, 700, 837], [500, 457, 360], [248, 342, 436]],
                labels = ['E700\nN=4', 'E400\nN=3', 'E250\nN=1'],
                patch_artist=True);

plt.ylabel('Width (nm)', fontsize= 'xx-large')
plt.xlabel('Pattern', fontsize= 'xx-large')
plt.xticks(fontsize= 'xx-large')
plt.yticks([200, 400, 600, 800], fontsize= 'xx-large')
plt.ylim(bottom=150, top = 950)
plt.grid()

bp['boxes'][0].set(color = 'purple', hatch = '/', linewidth=4, alpha = 0.3)
bp['boxes'][1].set(color = 'orange', hatch = '/', linewidth=4, alpha = 0.4)
bp['boxes'][2].set(color = 'green', hatch = '/', linewidth=4, alpha = 0.3)

bp['medians'][0].set(color = 'purple', linewidth=4, alpha = 0.5)
bp['medians'][1].set(color = 'orange', linewidth=4, alpha = 0.6)
bp['medians'][2].set(color = 'green', linewidth=4, alpha = 0.5)
    
plt.scatter(x = [1, 1, 1, 1], y = [896, 642, 700, 837], color='purple')
plt.scatter(x = [ 2, 2, 2], y = [500, 457, 360], color='orange')
plt.scatter(x = [3], y = [342], color='green')


plt.savefig("boxplot_width_modified")
```

### Quantification of the contrast

```python
plt.figure(figsize=(10,10))

bp = plt.boxplot([[0.267, 0.293, 0.356, 0.175], [0.044, 0.271, 0.03], [-0.045, 0.055, 0.155]],
                labels = ['E700\nN=4', 'E400\nN=3', 'E250\nN=1'],
                patch_artist=True);

plt.ylabel('Contrast', fontsize= 'xx-large')
plt.xlabel('Pattern', fontsize= 'xx-large')
plt.xticks(fontsize= 'xx-large')
plt.yticks([0.1, 0.2, 0.3, 0.4], fontsize= 'xx-large')
plt.ylim(bottom=-0.05, top = 0.45)
plt.grid()
    
bp['boxes'][0].set(color = 'purple', hatch = '/', linewidth=4, alpha = 0.3)
bp['boxes'][1].set(color = 'orange', hatch = '/', linewidth=4, alpha = 0.4)
bp['boxes'][2].set(color = 'green', hatch = '/', linewidth=4, alpha = 0.3)

bp['medians'][0].set(color = 'purple', linewidth=4, alpha = 0.5)
bp['medians'][1].set(color = 'orange', linewidth=4, alpha = 0.6)
bp['medians'][2].set(color = 'green', linewidth=4, alpha = 0.5)
    
plt.scatter(x = [1, 1, 1, 1], y = [0.267, 0.293, 0.356, 0.175], color='purple')
plt.scatter(x = [ 2, 2, 2], y = [0.044, 0.271, 0.03], color='orange')
plt.scatter(x = [3], y = [0.055], color='green')


plt.savefig("boxplot_contrast_modified")
```
